Job openings, separating, and hiring data was sourced from the "Job Openings and Labor Turnover Survey" (or JOLTS) from U.S. Bureau of Labor Statistics.

Note: Job openings data are divided into two types: 1) seasonally adjusted and 2) non-seasonally adjusted.

According to the documentations, **seasonally adjusted** data is altered by measuring and removing the influences of predictable seasonal patterns from time series data. It permits data over time to be compared across time.

More information here: https://www.bls.gov/help/one_screen/JT.htm#select-select-seasonal-adjustment

This notebook comprises both sets of data related to job openings.

```python
#importing necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

****Employment data from 2001-2023****

```python
#loading seasonally adjuseted data
HiresAdj = pd.read_csv("./data/hires-seasonally-adjusted.csv", skiprows=13) #hiring data
OpeningsAdj = pd.read_csv("./data/job-openings-seasonally-adjusted.csv", skiprows=13) #job opening data
SeparationAdj = pd.read_csv("./data/total-separations-seasonally-adjusted.csv", skiprows=13) #job separation data

HiresAdj.head()
#OpeningsAdj.head()
#SeparationAdj.head()
```

```python
#loading non-seasonally adjusted data
HiresNonAdj = pd.read_csv("./data/hires-non-adjusted.csv", skiprows=13)
OpeningsNonAdj = pd.read_csv("./data/job-openings-non-adjusted.csv", skiprows=13)
SeparationAdj = pd.read_csv("./data/separations-non-adjusted.csv", skiprows=13) 

HiresNonAdj.head()
#OpeningsNonAdj.head()
#SeparationNonAdj.head()
```

```python
HiresAdj["AnnualTotal"] = HiresAdj[HiresAdj.columns[1:]].sum(axis=1)
OpeningsAdj["AnnualTotal"] = OpeningsAdj[OpeningsAdj.columns[1:]].sum(axis=1)
SeparationAdj["AnnualTotal"] = SeparationAdj[SeparationAdj.columns[1:]].sum(axis=1)
```

```python
plt.plot(HiresAdj.Year, HiresAdj.AnnualTotal, label="Hires")
plt.plot(OpeningsAdj.Year, OpeningsAdj.AnnualTotal, label="Openings")
plt.plot(SeparationAdj.Year, SeparationAdj.AnnualTotal, label="Separation")
plt.xlabel("Years")
plt.ylabel("Frequency (levels)")
plt.title("U.S. Total Job Hiring, Openenings, & Separation 2000-2023")
plt.legend()
plt.show()
```

### Non-immigrant Workers data
H-1B data was sourced from USCIS's H-1B Employer Data Hub.(https://www.uscis.gov/tools/reports-and-studies/h-1b-employer-data-hub/understanding-our-h-1b-employer-data-hub)

It presents the data such that each row represents the number of petitions one employer filed for a foreign employee in the given fiscal year. 

***Key columns***:\
**Initial/Denial Approval** - number of approved/denied for foreign employees requiring H-1B visa to start an employment.\
**Continued Approval/Denial** - number of approved/denied for foreign employees requiring H-1B visa to continue an employment.

```python
#loading H1B data
H1B1 = pd.read_csv("H1B2023-2022.csv") #hiring data
```

```python
H1B1.head()
```

```python
H1B2 = pd.read_csv("H1B2021-2019.csv")
```

```python
H1B2.head()
```

```python
H1B3 = pd.read_csv("H1B2018-2016.csv")
```

```python
H1B3.head()
```

```python
H1B4 = pd.read_csv("H1B2015-2013.csv")
```

```python
H1B4.head()
```

```python
H1B5 = pd.read_csv("H1B2012-2009.csv")
```

```python
H1B5.head()
```

```python
H1B = pd.concat([H1B1, H1B2, H1B3, H1B4, H1B5], axis=0)
```

```python
len(H1B.index)
```

```python
H1B.columns = H1B.columns.str.replace(' ', '')
```

```python
H1B.head(10)
```

```python
H1B["InitialApproval"] = H1B["InitialApproval"].str.replace(",","")
```

```python
H1B.head(10)
```

```python
H1B["InitialApproval"] = H1B["InitialApproval"].astype(int)
```

```python
approved = H1B[["FiscalYear", "InitialApproval"]].groupby("FiscalYear", as_index=False)["InitialApproval"].sum()
```

```python
H1B["InitialDenial"] = H1B["InitialDenial"].astype(str)
```

```python
H1B.head()
```

```python
H1B["InitialDenial"] = H1B["InitialDenial"].str.replace(",","")
```

```python
H1B.head()
```

```python
H1B["InitialDenial"].unique()
```

```python
H1B["InitialDenial"] = H1B["InitialDenial"].astype(int)
```

```python
H1B.head()
```

```python
denied = H1B[["FiscalYear", "InitialDenial"]].groupby("FiscalYear", as_index=False)["InitialDenial"].sum()
```

```python
denied.head(15)
```

```python
H1B["totalInitial"] = H1B["InitialApproval"]+H1B["InitialDenial"]
```

```python
total = H1B[["FiscalYear", "totalInitial"]].groupby("FiscalYear", as_index=False)["totalInitial"].sum()
```

```python
plt.bar(total.FiscalYear,total.totalInitial, color='orange', label='denied', ec='black')
plt.bar(approved.FiscalYear, approved.InitialApproval, color='blue', label='approved',ec='black')
plt.xlabel("Years")
plt.ylabel("Approved H1B applications")
plt.title("Total H1B Petitions")
plt.legend()
plt.show()
```

```python
percentage = approved["InitialApproval"]/total["totalInitial"]
```

```python
percentage
```

```python
lmf = np.polyfit(total.FiscalYear, percentage,3)
xtheor = np.linspace(2009, 2023, 100)
ytheor= lmf[0]*xtheor**3 + lmf[1]*xtheor**2 + lmf[2]*xtheor + lmf[3]
plt.plot(xtheor,ytheor,color="red",label='LMF')

plt.bar(total.FiscalYear,percentage,ec='black')
plt.ylabel("Percentage Acceptance")
plt.xlabel("Years")
plt.title("Percentage H1B Acceptance")
plt.legend()
plt.show()
```

```python
industryApp = H1B[["Industry(NAICS)Code", "InitialApproval"]].groupby("Industry(NAICS)Code", as_index=False)["InitialApproval"].sum()
```

```python
industryApp.head()
```

```python
industryTot = H1B[["Industry(NAICS)Code", "totalInitial"]].groupby("Industry(NAICS)Code", as_index=False)["totalInitial"].sum()
```

```python
plt.bar(industryApp['Industry(NAICS)Code'], industryTot.totalInitial, color='orange')
plt.bar(industryApp['Industry(NAICS)Code'], industryApp.InitialApproval, color='blue')
plt.ylabel("Frequency")
plt.xlabel("Years")
plt.xticks(rotation = 90)
plt.title("Approval by Industry")
plt.show()
```

```python
labels = ["Farming, logging, fishing and hunting", "Natural Resources", "Utilities", "Construction", "Manufacturing", 
          "Wholesale", "Retail", "Transport & Storage", "Information", "Finance", "Real Estate", "STEM", "Management", 
         "Administration", "Education", "Healthcare", "Entertainment", "Accommodation", "Other", "Public Admin"]
patches, texts = plt.pie(industryTot.totalInitial, labels=None)
plt.legend(patches, labels, loc='center right', bbox_to_anchor=(0, 0.5),
           fontsize=8)
plt.title("Industry Applicant Proportion.")
plt.show()
```

```python
plt.bar(industryApp['Industry(NAICS)Code'], industryTot.totalInitial, color='orange', label='approved')
plt.bar(industryApp['Industry(NAICS)Code'], industryApp.InitialApproval, color='blue', label='denied')
plt.ylabel("Frequency")
plt.yscale("log")
plt.xlabel("Industry")
plt.xticks(rotation = 90)
plt.title("Approval by Industry")
plt.legend()
plt.show()
```

```python
HiresAdj["pct_change"] = HiresAdj["AnnualTotal"].pct_change()
OpeningsAdj["pct_change"] = OpeningsAdj["AnnualTotal"].pct_change()
SeparationAdj["pct_change"] = SeparationAdj["AnnualTotal"].pct_change()
approved["pct_change"] = approved["InitialApproval"].pct_change()
```

```python
approved["pct_change"] = approved["pct_change"].fillna(0)
```

```python
HiresAdj["pct_change"] = HiresAdj["pct_change"].fillna(0)
OpeningsAdj["pct_change"] = OpeningsAdj["pct_change"].fillna(0)
SeparationAdj["pct_change"] = SeparationAdj["pct_change"].fillna(0)
```

```python
lmf = np.polyfit(approved["pct_change"], HiresAdj["pct_change"][8:],1)
xtheor = np.linspace(min(approved["pct_change"]), max(approved["pct_change"]), 100)
ytheor= lmf[0]*xtheor + lmf[1]
correlation = round(approved["pct_change"].corr(HiresAdj["pct_change"][8:]),2)
plt.plot(xtheor,ytheor,color="red",linestyle='dashed',label='LMF, r='+str(correlation))

plt.scatter(approved["pct_change"],HiresAdj["pct_change"][8:])
plt.xlabel("Approval rate")
plt.ylabel("Hiring rate")
plt.title("Approval rate vs. Hiring rate.")
plt.legend()
plt.show()
```

```python
lmf = np.polyfit(approved["pct_change"], OpeningsAdj["pct_change"][8:],1)
xtheor = np.linspace(min(approved["pct_change"]), max(approved["pct_change"]), 100)
ytheor= lmf[0]*xtheor + lmf[1]
correlation = round(approved["pct_change"].corr(OpeningsAdj["pct_change"][8:]),2)
plt.plot(xtheor,ytheor,color="red",linestyle='dashed',label='LMF, r='+str(correlation))

plt.scatter(approved["pct_change"],OpeningsAdj["pct_change"][8:])
plt.xlabel("Approval rate")
plt.ylabel("Openings rate")
plt.title("Approval rate vs. Openings rate.")
plt.legend()
plt.show()
```

```python
lmf = np.polyfit(approved["pct_change"], SeparationAdj["pct_change"][8:],1)
xtheor = np.linspace(min(approved["pct_change"]), max(approved["pct_change"]), 100)
ytheor= lmf[0]*xtheor + lmf[1]
correlation = round(approved["pct_change"].corr(SeparationAdj["pct_change"][8:]),2)
plt.plot(xtheor,ytheor,color="red",linestyle='dashed',label='LMF, r='+str(correlation))

plt.scatter(approved["pct_change"],SeparationAdj["pct_change"][8:])
plt.xlabel("Approval rate")
plt.ylabel("Separations rate")
plt.title("Approval rate vs. Separations rate.")
plt.legend()
plt.show()
```

### Packing the data

```python
packed = pd.DataFrame()
```

```python
packed["year"] = np.arange(2009, 2024)
```

```python
packed["hiring"] = HiresAdj['AnnualTotal'][8:].reset_index(drop=True)
packed["separation"] = SeparationAdj['AnnualTotal'][8:].reset_index(drop=True)
packed["opening"] = OpeningsAdj['AnnualTotal'][8:].reset_index(drop=True)
```

```python
packed["H1B_approved"] = approved['InitialApproval']
packed["H1B_denied"] = denied['InitialDenial']
packed["H1B_total"] = denied['InitialDenial'] + approved['InitialApproval']
```
