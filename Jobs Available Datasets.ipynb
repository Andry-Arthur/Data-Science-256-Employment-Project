The data was sourced from the "Job Openings and Labor Turnover Survey" (or JOLTS) from U.S. Bureau of Labor Statistics.

Note: Job openings data are divided into two types: 1) seasonally adjusted and 2) non-seasonally adjusted.

According to the documentations, **seasonally adjusted** data is altered by measuring and removing the influences of predictable seasonal patterns from time series data. It permits data over time to be compared across time.

More information here: https://www.bls.gov/help/one_screen/JT.htm#select-select-seasonal-adjustment

This notebook comprises both sets of data related to job openings.
